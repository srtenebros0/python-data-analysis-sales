IMPORTAMOS LIBRERIAS Y HARDCODEAMOS LA DATA

```python
import pandas as pd

data = {
    "vendedor": ["Ana","Luis","Carlos","Maria","Ana","Luis","Carlos","Maria"],
    "ventas": [200,500,150,300,400,250,100,350],
    "mes": ["Enero","Enero","Enero","Enero","Febrero","Febrero","Febrero","Febrero"]
}

df = pd.DataFrame(data)
df
```

Miramos la informacion de los campos para saber su tipo de dato

```python
df.info()
```

Sacamos el total de ventas, promedio y total de registros

```python
prom_vent = df.groupby("vendedor")["ventas"].agg(["sum","mean","count"])
prom_vent
```

Creamos funcion de clasificacion de ventas por vendedor

```python
def TopVentas(ventas):
  if ventas >= 350:
    return "Top"
  if ventas >= 200:
    return "Medio"
  if ventas < 200:
    return "Bajo"
```

Creamos columna de Categorias y mandamos a llamar la funcion de TopVentas para la clasisficacion.

```python
df["Categoria"] = df["ventas"].apply(TopVentas)
df
```

Creamos columna de bonos y asignamos bono a cada vendedor

```python
df["Bono"] = df["ventas"] * 0.50
df
```

```python
showBonos= df.sort_values(by="ventas", ascending=False)
showBonos
```

```python
mejor_vendedor = df.sort_values(by="ventas", ascending=False).iloc[0]
peor_vendedor = df.sort_values(by="ventas", ascending=True).iloc[0]
#print(f"{mejor_vendedor}------{peor_vendedor}")
```

```python
ventas_por_mes = df.groupby("mes")["ventas"].sum()
mejor_mes = ventas_por_mes.idxmax()
peor_mes = ventas_por_mes.idxmin()

mejor_mes_venta = ventas_por_mes.max()
peor_mes_venta = ventas_por_mes.min()

print(f"{mejor_mes} - {mejor_mes_venta}")
print(f"{peor_mes} - {peor_mes_venta}")
```

```python
print(f"""
Conclusin final del analisis de ventas de los meses de Enero y Febrero : \n
Tenemos a {mejor_vendedor["vendedor"]} destacando en la tabla de ventas con un total de {mejor_vendedor["ventas"]}\n siendo {mejor_vendedor["mes"]} el mes que mas ventas registro.\n
Por el lado contrario tenemos a {peor_vendedor["vendedor"]} con un total de ventas de {peor_vendedor["ventas"]}\n siendo el mes de {peor_vendedor["mes"]} en el que menos ventas registro.\n Por lo que automaticamente {peor_vendedor["vendedor"]} es quien mas necesita apoyo de capacitacion. \n

Como resumen en general, tuvimos el mes de {mejor_mes} con un total de {mejor_mes_venta} de ventas\n y el mes de {peor_mes} con un total de {peor_mes_venta} de ventas.

Se recomienda asignar capacitacion a {peor_vendedor["vendedor"]} para mejorar su rendimiento en la proxima evaluacion bimestral.

A continuacin dejo la tabala completa como respaldo de los datos, donde vienen las ventas y bonos de cada vendedor.

{df}

""")
```
